# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

__Ans. :__<br>

__1. Scatter plot:__ This plot is useful for visualizing the relationship between two continuous variables. It can be used to identify patterns, trends, or outliers in the data.

__2. Histogram:__ This plot is useful for visualizing the distribution of a single variable. It can be used to identify the frequency of values in a given range or bin.

__3. Bar plot:__ This plot is useful for visualizing the relationship between a categorical variable and a numerical variable. It can be used to compare the values of different categories or subgroups.

__4. Box plot:__ This plot is useful for visualizing the distribution of a numerical variable across different categories or subgroups. It can be used to identify outliers, variability, and skewness in the data.

__5. Heatmap:__ This plot is useful for visualizing the relationship between two categorical variables. It can be used to identify patterns or trends in the data, such as correlations or associations between different categories or subgroups.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

__Note:__ timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
fmri_data = sns.load_dataset("fmri")
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.lineplot(x=fmri_data.timepoint, y=fmri_data.signal, data=fmri_data, style='event', hue='region')
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

__Note:__ pclass, age, and fare are columns in the titanic dataset.

```python
# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

sns.set(rc={'figure.figsize':(11.7,8.27)})
# Plot a box plot for age, grouped by passenger class
sns.boxplot(data=titanic_data, x="pclass", y="age")

# Plot a box plot for fare, grouped by passenger class
sns.boxplot(data=titanic_data, x="pclass", y="fare")
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds_data = sns.load_dataset("diamonds")
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.histplot(data=diamonds_data, x="price", hue="cut", multiple="stack")
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
# Load the iris dataset
iris_data = sns.load_dataset("iris")

sns.set(rc={'figure.figsize':(11.7,8.27)})

# Plot a pair plot with different colors for each species
sns.pairplot(data=iris_data, hue="species")
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Reshape the data to create a matrix with months on one axis and years on the other
flights_matrix = flights_data.pivot("month", "year", "passengers")

sns.set(rc={'figure.figsize':(11.7,8.27)})

# Plot a heatmap of the matrix
sns.heatmap(data=flights_matrix, cmap="coolwarm", annot=True, fmt="d")
```
